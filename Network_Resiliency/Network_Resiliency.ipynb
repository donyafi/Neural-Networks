```python
import networkx
import networkx.algorithms.centrality as Centrality
import networkx.algorithms.connectivity as Connectivity
import networkx.algorithms.components as Components
import matplotlib.pyplot as plt
import random
import operator


class Network_Resiliency_Test:
    
    def __init__(self):
        print("Net Created")
        self.Network = networkx.Graph()

    def LoadDataset(self,Path):
        self.Network=networkx.read_edgelist(Path)
    
    def StoreGraph(self,Path):
        networkx.write_edgelist(self.Network,Path)
        print("network is stored")
        
    def DrawNetwork(self):
        networkx.draw_networkx(self.Network)
        plt.figure(figsize=(50,50))
        plt.show() 

    def Remove_Node(self,nid):
        try:
            self.Network.remove_node(str(nid))
            print('node', nid , 'removed')
        except:
            print("not found " + str(nid))
            pass 

    def Remove_Edge(self,src,dst):
        self.Network.remove_edge(src,dst)

    def Average_ShortestPath(self):
        try:
            return networkx.average_shortest_path_length(self.Network)
        except:
            return -1
    
    def Diameter(self):
        try:
            return networkx.diameter(self.Network)
        except:
            return -1
    
    def DegreeCentrality(self):
        return Centrality.degree_centrality(self.Network)

    def ClosenessCentrality(self):
        return Centrality.closeness_centrality(self.Network)

    def BetweennessCentrality(self):
        betweenness=Centrality.betweenness_centrality(self.Network)
        return betweenness

    def Connectability(self):
        return Connectivity.connectivity.average_node_connectivity(self.Network) #, Connectivity.all_pairs_node_connectivity(self.Network)

    def ConnectedComponents(self):
        Connected_Components = Components.connected_components(self.Network)
        for c in sorted(Connected_Components, key=len, reverse=True):
            print("Connected Component Size is : " + str(len(c)))

    def Approaximate_Persistance(self):
        betweenness=objct.BetweennessCentrality()
        sorted_tuples = sorted(betweenness.items(), key=operator.itemgetter(1), reverse=True)
        # print(sorted_tuples)
        sorted_dict = {k: v for k, v in sorted_tuples}
        maxDiameter=6
        removed_nodes_count=-1
        for key in sorted_dict:
            if objct.Diameter()>=maxDiameter:
                objct.Remove_Node(key)
                removed_nodes_count+=1
                print(objct.Diameter())
            else:
                print('last Diameter:',objct.Diameter())
                print('number of removed nodes:', removed_nodes_count)
                break
            
objct=Network_Resiliency_Test()
objct.LoadDataset("graph_data.txt")
```
